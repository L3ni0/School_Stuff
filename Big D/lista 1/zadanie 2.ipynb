```python
from matplotlib import pyplot as plt
import numpy as np
import random
```

```python
X = np.linspace(0.0,10.0,30)
Y = (X**2 + X*2 +2)
Y = [y+ random.uniform(-2,2) for y in Y]
```

```python
model = np.poly1d(np.polyfit(X,Y,2))
plt.scatter(X,Y, color='green')
plt.plot(X, model(X), color='red')
display(model.coef)
plt.show()
```
