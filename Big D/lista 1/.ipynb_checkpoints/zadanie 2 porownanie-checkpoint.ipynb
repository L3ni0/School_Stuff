```python
from matplotlib import pyplot as plt
from scipy import optimize
import numpy as np
import random
```

```python
X = np.linspace(0.0,20.0,60)
dane = [3,-4,-1]
def model(x, dane):
    a,b,c = dane
    return x**a + x*b + c
Y = model(X, dane)
Y = np.array([y + random.uniform(-1,2) for y in Y])
Y
```

```python
regresja = np.poly1d(np.polyfit(X,Y,2))
display(regresja.coef)
plt.scatter(X,Y, label='wartości rzeczywite')
Y_numpy = model(X, regresja.coef)
plt.plot(X, Y_numpy, color='green', label='wartości predykowane')
plt.title('regresja liniowej za pomocą biblioteki numpy')
plt.xlabel('wartości X')
plt.ylabel('wartości Y')
plt.legend()
plt.show()
```

```python
def blad_kwadratowy(dane):
    y_pred = model(X, dane)
    return ((Y-y_pred)**2).mean()



estymatowana= optimize.fmin(blad_kwadratowy, np.array([0,0,0]))
Y_optimize = model(X, estymatowana)
plt.scatter(X,Y, label='wartości rzeczywiste')
plt.plot(X,Y_optimize, label='wartości obliczone wykorzystując optymalizacje', color='green')
plt.ylabel('wartości Y')
plt.xlabel('wartości X')
plt.legend()
plt.show()
```

```python
plt.plot(X, Y_numpy, color='red', label='wartości predykowane')
plt.title('regresja liniowej za pomocą biblioteki numpy')
plt.plot(X,Y_optimize, label='wartości obliczone wykorzystując optymalizacje', color='green')
plt.scatter(X,Y, label='wartości rzeczywite')
plt.xlabel('wartości X')
plt.ylabel('wartości Y')
plt.legend()
```
