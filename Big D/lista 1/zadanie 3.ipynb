```python
import matplotlib.pyplot as plt
import numpy
import numpy as np
from sklearn.linear_model import Ridge
import random
from scipy import optimize
```

```python
X = np.linspace(0,20,60)
def model(dane, X):
    a,b =dane
    return a*X +b

dane = [2,4]
Y = np.array([model(dane,x) + random.uniform(-7,7) for x in X])
X = [[x] for x in X]
plt.scatter(X,Y)
```

```python
ridge_reg = Ridge(2)
Y.transpose()

ridge_reg.fit(X,Y)
y_pred = ridge_reg.predict(X)
plt.scatter(X,Y, label='punkty rzeczywiste', color='green')
plt.plot(X, y_pred, label='punkty predykowane')
plt.legend()
plt.show()
```
