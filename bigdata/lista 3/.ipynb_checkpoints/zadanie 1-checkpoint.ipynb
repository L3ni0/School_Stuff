```python

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf
```

```python
X = np.arange(0,1000)
generated_data = ArmaProcess(np.array([1, -0.9]),np.array([1])).generate_sample(len(X))
plt.figure(1, [20,5])
plt.plot(X,generated_data)
```

```python
train_Y = generated_data[:-100]
model_ar = AutoReg(train_Y,1)
mod = model_ar.fit()
```

```python
plot_acf(generated_data, alpha=1, lags=20)
```

```python
mod.plot_predict(start=100, end=1000)
```
